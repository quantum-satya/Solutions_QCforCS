### Write three functions that perform the addition, inverse, and scalar multiplication operations for Cn, i.e., write a function that accepts the appropriate input for each of the operations and outputs the vector.

```python
# Write three functions that perform the addition, inverse, and scalar multiplication
# operations for Cn, i.e., write a function that accepts the appropriate input for each of the operations and outputs the vector.
import numpy as np  
def vector_addition(v1, v2):
    return np.array(v1) + np.array(v2)  
def vector_inverse(v):
    return -np.array(v)  
def scalar_multiplication(scalar, v):
    return scalar * np.array(v)  
# Example usage
v1 = [1+2j, 3+4j, 5+6j]
v2 = [6+5j, 4+3j, 2+1j]
scalar = 2 
print("Vector Addition:", vector_addition(v1, v2))
print("Vector Inverse of v1:", vector_inverse(v1))
print("Scalar Multiplication of v1 by", scalar, ":", scalar_multiplication(scalar, v1))
#!/usr/bin/env python3
```
