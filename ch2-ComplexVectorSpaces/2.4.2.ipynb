### Write a function that calculates the norm of a given complex vector.

```python
# Write a function that calculates the norm of a given complex vector.
import numpy as np
def vector_norm(vector):    
    return np.linalg.norm(vector)   
# Example usage
vector = np.array([3+4j, 1-1j, -2+2j])
norm = vector_norm(vector)
print("Vector:", vector)
print("Norm of the vector:", norm)
```
