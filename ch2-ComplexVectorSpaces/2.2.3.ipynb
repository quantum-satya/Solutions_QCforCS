### Write a function that accepts a vector and a matrix and outputs the vector resulting fron the "action" 

```python
# Writa a function that accepts a vector and a matrix and outputs the vector resulting fron the "action"
import numpy as np
def matrix_vector_action(matrix, vector):
    return np.dot(matrix, vector)
# Example usage
matrix = np.array([[1+1j, 2+2j], [3+3j, 4+4j]])
vector = np.array([1+0j, 0+1j])
result_vector = matrix_vector_action(matrix, vector)
print("Matrix:\n", matrix)
print("Vector:\n", vector)
print("Resulting Vector after Action:\n", result_vector) 

```
