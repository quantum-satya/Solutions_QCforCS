### Write a function that accepts a square matrix and tells if it is hermitian.

```python
# Write a function that accepts a square matrix and tells if it is hermitian.
import numpy as np
def is_hermitian(matrix):
    return np.array_equal(matrix, matrix.conj().T)
# Example usage
matrix = np.array([[2, 2-1j], [2+1j, 3]])
print("Matrix:\n", matrix)
if is_hermitian(matrix):
    print("The matrix is Hermitian.")
else:
    print("The matrix is not Hermitian.")
```
