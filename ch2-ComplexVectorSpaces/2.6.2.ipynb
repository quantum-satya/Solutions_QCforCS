### Write a function that accepts a square matrix and tells if it is unitary.

```python
# Write a function that accepts a square matrix and tells if it is unitary.
import numpy as np
def is_unitary(matrix):
    identity = np.eye(matrix.shape[0])
    product = np.dot(matrix, matrix.conj().T)
    return np.array_equal(product, identity)
# Example usage
matrix = np.array([[1/np.sqrt(2), 1/np.sqrt(2)], [1/np.sqrt(2), -1/np.sqrt(2)]])
print("Matrix:\n", matrix)
if is_unitary(matrix):
    print("The matrix is Unitary.")
else:
    print("The matrix is not Unitary.")
    
```
