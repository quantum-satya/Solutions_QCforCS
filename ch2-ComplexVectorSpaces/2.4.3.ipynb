### Write a function that calculates the distance of two given complex vectors.

```python
# Write a function that calculates the distance of two given complex vectors.
import numpy as np
def vector_distance(vec1, vec2):
    return np.linalg.norm(np.array(vec1) - np.array(vec2))
# Example usage
vec1 = np.array([1+2j, 3+4j, 5+6j])
vec2 = np.array([6+5j, 4+3j, 2+1j])
distance = vector_distance(vec1, vec2)
print("Vector 1:", vec1)
print("Vector 2:", vec2)
print("Distance between the vectors:", distance)
```
