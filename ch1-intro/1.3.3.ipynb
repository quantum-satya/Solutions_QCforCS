### Expand your program. Add functions for multiplication, division, and rerurning the polar coordinates of a number.

```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Prompt for a complex number
    z1 = complex(input("Enter a complex number (e.g. 1+1j): ").strip())
    z2 = complex(input("Enter another complex number for operations (e.g. 2+2j): ").strip())
    print(f"You entered z1 ={z1} and z2 = {z2}")

    # Functions for multiplication, division, and polar coordinates
    def multiply(z1, z2):
        return z1 * z2

    def divide(z1, z2):
        return z1 / z2

    def to_polar(z1):
        r = abs(z1)
        theta = np.angle(z1)
        return r, theta

    # Example usage
    
    product = multiply(z1, z2)
    quotient = divide(z1, z2)
    r, theta = to_polar(z1)

    print(f"Product: {product}")
    print(f"Quotient: {quotient}")
    print(f"Polar coordinates: r = {r}, theta = {theta} radians")    

    # Plot the vector z1
    # Create a figure and axis
    #plt.figure()
    
    # Create a polar plot
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.set_title("Complex vector in (r, ) form", pad=20) 

    # Plot the vector: start at origin, end at (r, )
    ax.arrow(0, 0, theta, r, head_width=0.1, head_length=0.2,fc='blue', ec='blue')   

    # Annotate the point with its (r, ) values
    ax.text(theta, r, f"({r:.2f}, {theta:.2f} rad)",ha='left', va='bottom', fontsize=10) 

    plt.show() 

if __name__ == "__main__":
    main()
```
