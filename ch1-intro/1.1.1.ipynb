### Write a python program that accepts two complex numbers and outputs their sum and their product.

```python
def main():
    # Prompt for two complex numbers
    z1 = complex(input("Enter first complex number (e.g. 3+4j): ").strip())
    z2 = complex(input("Enter second complex number (e.g. 1-2j): ").strip())

    # Compute sum and product
    s = z1 + z2
    p = z1 * z2

    # Display the results
    print(f"Sum    : {s}")
    print(f"Product: {p}")

if __name__ == "__main__":
    main()
```

## Let's break it down line by line:


def main():

- This line defines a function named main. This is the entry point of the program, where execution begins.


z1 = complex(input("Enter first complex number (e.g. 3+4j): ").strip())

- input(...) prompts the user to enter a complex number and reads the input as a string.
- .strip() removes any leading or trailing whitespace from the input string.
- complex(...) converts the input string to a complex number, which is stored in z1.


z2 = complex(input("Enter second complex number (e.g. 1-2j): ").strip())

- This line does the same as the previous one, but for the second complex number, storing it in z2.


s = z1 + z2

- This line calculates the sum of z1 and z2 using the + operator, which is overloaded for complex numbers.
- The result is stored in s.


p = z1 * z2

- This line calculates the product of z1 and z2 using the * operator, which is overloaded for complex numbers.
- The result is stored in p.


print(f"Sum    : {s}")
print(f"Product: {p}")

- These lines print the sum and product to the console, using f-strings to format the output.
- The f prefix allows us to embed expressions (s and p) inside the string.


if __name__ == "__main__":

- This line checks if the script is being run directly (not being imported as a module).
- __name__ is a built-in variable that holds the name of the current module ("__main__" when run directly).


main()

- If the script is being run directly, this line calls the main function, starting the program's execution.

Overall, this program reads two complex numbers, calculates their sum and product, and prints the results.
