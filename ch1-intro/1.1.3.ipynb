## PD 1.1.3: 
###  Write a program that converts a complex number from its Cartesian representation to its 
### polar representation and vice versa. 

```python
import numpy as np

def main():
    # Prompt for a complex number
    z = complex(input("Enter a complex number (e.g. 3+4j): ").strip())

    print(f"You entered z = {z}")

    # Convert to polar representation
    r = abs(z)
    theta = np.angle(z)

    print(f"Polar representation: r = {r}, theta = {theta} radians")

    # Convert back to Cartesian representation
    z_cartesian = r * (np.cos(theta) + 1j * np.sin(theta))

    print(f"Back to Cartesian: {z_cartesian}")
if __name__ == "__main__":
    main()
 


```
