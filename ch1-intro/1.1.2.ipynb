### make it also perform subtraction and division of complex numbers. In addition, let the user enter a complex 
### number and have the computer return its modulus and conjugate.

```python
def main():
    # Prompt for two complex numbers
    z1 = complex(input("Enter first complex number (e.g. 3+4j): ").strip())
    z2 = complex(input("Enter second complex number (e.g. 1-2j): ").strip())

    print(f"You entered z1 = {z1} and z2 = {z2}")


    # Compute sum and product
    sum = z1 + z2
    prod = z1 * z2
    diff = z1 - z2
    quot = z1 / z2

    # Compute modulus and conjugate
    mod_z1 = abs(z1)
    conj_z1 = z1.conjugate()
    mod_z2 = abs(z2)
    conj_z2 = z2.conjugate()    
    # Display the results
    
    print(f"Sum         : {sum}")
    print(f"Product     : {prod}")
    print(f"Difference  : {diff}")
    print(f"Quotient    : {quot}")
    print(f"Modulus z1  : {mod_z1}")
    print(f"Conjugate z1: {conj_z1}")
    print(f"Modulus z2  : {mod_z2}")
    print(f"Conjugate z2: {conj_z2}")   

if __name__ == "__main__":
    main()
```
